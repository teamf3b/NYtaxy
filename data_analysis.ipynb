```r
a = read.csv("data/train.csv")
```

```r
(head(a, n=10))
```

```r
passengers = table(a$passenger)

barplot(passengers, log="y")
```

```r
hist(subset(a$trip_duration, a$trip_duration < 5000))
```

```r
library(ggplot2)
b = subset(a, a$trip_duration < 10000)
b$passenger_count = as.factor(b$passenger_count)
ggplot(b, aes(x=passenger_count, y=trip_duration)) + geom_violin()
boxplot(trip_duration~passenger_count, data=b)
```

```r
ggplot(data=a, aes(x=pickup_longitude, y=pickup_latitude) ) +
  geom_bin2d(bins = 100) +
  theme_bw()
```

```r
c =  subset(a, a$pickup_longitude > -90 & a$pickup_latitude < 45)

ggplot(data=c, aes(x=pickup_longitude, y=pickup_latitude) ) +
  geom_hex(bins = 200) +
  theme_bw()
```
